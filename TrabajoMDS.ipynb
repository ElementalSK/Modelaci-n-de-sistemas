```python
#Codigo

import pandas as pd
import numpy as np

# Cargar los datos
motivacion_df = pd.read_csv("Cuestionario motivacion academica.csv")
col_reprobadas = 'indica la cantidad de asignaturas reprobadas desde su inicio de la carrera hasta la fecha. si no has reprobado, marca 0'
col_motivacion = 'indica tu nivel actual de motivaci√≥n por estudiar tu carrera'

# Verificar que las columnas existan
if col_reprobadas in motivacion_df.columns and col_motivacion in motivacion_df.columns:

    # Calcular puntuaci√≥n de riesgo
    motivacion_df["reprob_predicha"] = (
        motivacion_df[col_reprobadas] * 1.5
        - motivacion_df[col_motivacion] * 0.5
    )

    # Ajustar valores negativos
    motivacion_df["reprob_predicha"] = motivacion_df["reprob_predicha"].clip(lower=0)

    # Definir percentiles para clasificaci√≥n
    p_bajo = np.percentile(motivacion_df["reprob_predicha"], 70)
    p_medio = np.percentile(motivacion_df["reprob_predicha"], 85)

    # Funci√≥n para determinar nivel de alerta
    def nivel_alerta(reprob_predicha):
        if reprob_predicha <= p_bajo:
            return "üü¢ Bajo riesgo"
        elif reprob_predicha <= p_medio:
            return "üü† Riesgo medio"
        else:
            return "üî¥ Alto riesgo"

    # Aplicar la clasificaci√≥n
    motivacion_df["nivel_alerta"] = motivacion_df["reprob_predicha"].apply(nivel_alerta)

    # Guardar resultados
    output_path = "Analisis_Riesgo_Motivacion.csv"
    motivacion_df.to_csv(output_path, index=False, encoding='utf-8-sig')

    # Mostrar resultados
    print(f"Archivo guardado: {output_path}")
    print("Resumen de alertas generadas:")
    print(motivacion_df["nivel_alerta"].value_counts())

else:
    print("Error: No se encontraron las columnas necesarias en el dataset")
    print("Columnas disponibles:", motivacion_df.columns.tolist())
```

#Explicaci√≥n del metodo

La formula que desarrollamos contiene dos variables claves, estas son:

- Cuantas materias ha reprobado el estudiante.
- Que tan motivado se siente con su carrera.

Para aquello creamos una formula:

puntuacion_riesgo = (reprobadas * 1.5) - (motivacion * 0.5)

Esta formula significa que:

1.- Si un alumno ha reprobado varias materias, eso indica m√°s preocupaci√≥n ‚Üí por eso se multiplicamos por 1.5
2.- Si el alumno est√° motivado, eso puede ayudar a mejorar su condici√≥n academica ‚Üí por eso se le resta la motivaci√≥n multiplicada por 0.5
3.- Al primer resultado se le resta el segundo.
- Si el valor es negativo, se deja en cero (no tiene sentido tener "riesgo negativo")

Clasificaci√≥n de los resultados:

¬∑ Los que est√°n en el 70% m√°s bajo ‚Üí üü¢ Bajo riesgo
¬∑ Del 70% al 85% ‚Üí üü† Riesgo medio
¬∑ Del 85% hacia arriba ‚Üí üî¥ Alto riesgo

B√°sicamente esto es:

- Un alumno con muchas reprobadas Y poca motivaci√≥n = alto riesgo.
- Un alumno con pocas reprobadas Y mucha motivaci√≥n = bajo riesgo.
- Los casos intermedios = riesgo medio.

La finalidad de esta formula es ver que si un estudiante ha reprobado varias materias y adem√°s est√° desmotivado, probablemente necesita m√°s apoyo.
Nuestra f√≥rmula detecta estos casos y los clasifica en verde, amarillo o rojo seg√∫n qu√© tan grave sea la situaci√≥n.
